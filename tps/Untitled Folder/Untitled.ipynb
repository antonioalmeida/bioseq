```python
def dot_plot(seq1, seq2):
    return [[1 if seq1[j] == seq2[i] else 0 for i in range(0, len(seq2))] for j in range(0, len(seq1))]
```

```python
dot_plot('ATCG', 'ATCGA')
```

```python
def pretty_header(seq):
    print('    ', end="")
    for i in seq:
        print(i + ' | ', end="")
    print()

def pretty_row(value, row):
    print(value + ' | ', end="")
    for i in row:
        print(str(i) + ' | ', end="")
    print()

def pretty_dot_plot(seq1, seq2):
    pretty_header(seq2)
    dp = dot_plot(seq1, seq2)
    
    for i,v in enumerate(seq1):
        pretty_row(str(v), dp[i])
```

```python
# Implements a substitution matrix given an alphabet, a value for match and a value for a mismatch

def subst_matrix(alphabet, match=1, mismatch=0):
    dic = { (x,y):match if x==y else mismatch for x in alphabet for y in alphabet}
    return dic
```

```python
# In practice for protein substitution matrix it is ieasier to read from a table or an existing file
# Assume the matrix is symmetric and first row contains the symbols in the alphabet
# Tip: use blosum62.mat

def read_submat_file(filename):
    
    with open(filename) as f:
        
        chars = f.readline().strip().replace('\t','')
    
        keys = [x for x in chars]
        matrix = []
        
        for line in f: 
            matrix.append(line.strip().split())
        
        dic = { x+y:int(matrix[yi][xi]) for xi,x in enumerate(keys) for yi,y in enumerate(keys) }
        return dic
```

```python
read_submat_file('blosum62.mat')
```

```python
# Provides de score of a column alignment, i.e., between characters c1 and c2
# Assume a constant gap penalty g and a substitution matrix sm

def score_col_alignment(c1, c2, sm, g):
    return g if c1 == '-' or c2 == '-' else sm[c1+c2] 
```

```python
sm = read_submat_file('blosum62.mat')
score_col_alignment('A','H', sm, 4)
```

```python
# Provides the score for sequece aligment
def score_seq(seq1, seq2, sm, g):
    assert len(seq1) == len(seq2), 'ERROR: different sized sequences'
    
    score = 0
    for i in range(0,len(seq1)):
        score+= score_col_alignment(seq1[i], seq2[i], sm, g)
        
    return score
```

```python
sm = read_submat_file('blosum62.mat')
assert score_seq('ARN-AKF', 'AR-ALKF', sm, -10) == -1, "should be -1"
assert score_seq('AR-ALF', 'AR-ALF', sm, -15) == 8, "should be 8"
```
